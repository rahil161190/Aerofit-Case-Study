## Feature Column Creation

```python
def place2state(x):
    # transform  "gurgaon_bilaspur_hb (haryana)" into "haryana)""
    state = x.split('(')[1] # haryana)

    return state[:-1] #removing ')' from ending # haryana

def place2city(x):
    # We will remove state #"gurgaon_bilaspur_hb (haryana)"
    city = x.split(' (')[0]#"gurgaon_bilaspur_hb

    city = city.split('_')[0]#"gurgaon

    #Now dealing with edge cases

    if city == 'pnq vadgaon sheri dpc':
      return 'vadgaonsheri'

    # ['PNQ Pashan DPC', 'Bhopal MP Nagar', 'HBR Layout PC',
    #  'PNQ Rahatani DPC', 'Pune Balaji Nagar', 'Mumbai Antop Hill']

    if city in ['pnq pashan dpc','pnq rahatani dpc', 'pune balaji nagar']:
        return 'pune'

    if city == 'hbr layout pc' : return 'bengaluru'
    if city == 'bhopal mp nagar' : return 'bhopal'
    if city == 'mumbai antop hill' : return 'mumbai'

    return city

def place2city_place(x):

    # We will remove state
    x = x.split(' (')[0]

    len_ = len(x.split('_'))

    if len_ >= 3:
        return x.split('_')[1]

    # Small cities have same city and place name
    if len_ == 2:
        return x.split('_')[0]


    # Now we need to deal with edge cases or imporper name convention

    #if len(x.split(' ')) == 2:
    #

    return x.split(' ')[0]


def place2code(x):
    # We will remove state
    x = x.split(' (')[0]

    if len(x.split('_')) >= 3 :
        return x.split('_')[-1]

    return 'none'
```

```python
df['Source_city'] = df['source_name'].apply(lambda x: place2city(x))
df['Source_state'] = df['source_name'].apply(lambda x: place2state(x))
df['Source_place'] = df['source_name'].apply(lambda x: place2city_place(x))
df['Source_code'] = df['source_name'].apply(lambda x: place2code(x))
```

```python
df['Destination_city'] = df['destination_name'].apply(lambda x: place2city(x))
df['Destination_state'] = df['destination_name'].apply(lambda x: place2state(x))
df['Destination_place'] = df['destination_name'].apply(lambda x: place2city_place(x))
df['Destination_code'] = df['destination_name'].apply(lambda x: place2code(x))
```

```python
from scipy import stats
from scipy.stats import f_oneway
from scipy.stats import ttest_ind
from scipy.stats import shapiro
from scipy.stats import levene
from scipy.stats import chi2_contingency
```

---


### Q-> 1 - Is the average actual time taken by trips are significantly different than osrm time ?
---



*   H0 is null hypothesis
*   Ha is alternate hypothesis



H0:  The average time taken by trips are same as predicted by OSRM time.

Ha: The average time taken by trips are not same as predicted by OSRM time.

Let Î¼1 and Î¼2 be the average time taken(actual) and average time taken(OSRM)respectively.

Mathematically, the above formulated hypothesis can be written as:

H0:Î¼1>=Î¼2

Ha:Î¼1!=Î¼2

```python
alpha = 0.05
```

# Kruskal Test

```python
#H0 Null hypothesis trip_actual_time_HT, trip_OSRM_time_HT are normally distributed
#Ha Alternate hypothesis trip_actual_time_HT, trip_OSRM_time_HT are  not normally distributed
from scipy.stats import kruskal
stat, p_value = kruskal(trip_actual_time_HT, trip_OSRM_time_HT)

print("test statistic:",stat)
print("p_value:",p_value)
result(p_value, alpha)
```

Observation: Since the p-value is lesser than the 5% significance level for both Kruskal wallis test and ttest, we will reject the null hypothesis. Hence, we have enough statistical evidence to say that the avg time (actual) and avg time(osrm) distribution is not normal.
